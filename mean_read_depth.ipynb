```python
import pandas as pd
import numpy as np
from natsort import natsort_keygen
import glob
from tqdm import tqdm
```

```python
#https://www.biostars.org/p/383786/
arms = pd.read_csv('chrom_arms.tsv', sep='\t')
chr_names = [f'chr{i}' for i in range(1, 23)]

# Keep only the chromosomes
arms = arms[arms['chrom'].isin(chr_names)]

# Sort the dataframe by chromosome and start position using natsort
arms.sort_values(by=['chrom', 'start_pos'], key=natsort_keygen(), inplace=True)

# Reset the index
arms.reset_index(drop=True, inplace=True)

# Convert start_pos, end_pos and lenght to integers
arms['start_pos'] = arms['start_pos'].astype(int)
arms['end_pos'] = arms['end_pos'].astype(int)
arms['length'] = arms['length'].astype(int)

arms
```
